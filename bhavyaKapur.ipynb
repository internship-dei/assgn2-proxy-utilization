```python
from lxml.html import fromstring
import requests
from itertools import cycle
import traceback
done = 0
no = 0
def get_proxies():
  url = 'https://free-proxy-list.net/'
  response = requests.get(url)
  parser = fromstring(response.text)
  proxies = set()
  for i in parser.xpath('//tbody/tr')[:10000000]:
    if i.xpath('.//td[7][contains(text(),"yes")]'):
      proxy = ":".join([i.xpath('.//td[1]/text()')[0], i.xpath('.//td[2]/text()')[0]])
      proxies.add(proxy)
  return proxies
proxies = get_proxies()
print(len(proxies))
print(proxies)

```

```python
proxy_pool = cycle(proxies)
url = "https://top10songs.com/"
for i in range(1,77):
  proxy = next(proxy_pool)
  print("Request #%d"%i)
  try:
    response = requests.get(url,proxies={"http": proxy, "https": proxy})
    print(response.status_code)
    done = done + 1
  except:
    print("Skipping. Connnection error")
    no = no + 1
    
sum = no + done
percentage = (done/sum)*100
print()
print(percentage)
```

```python
print("->Observations:")
print()
print("(a). Total no. of iterations = ", i)
print("(b). No. of successful iterations = ", done)
print("(c). No. of failed iterations = ", no)
print("(d). Success Hit Ratio = ", percentage)
```
